```python
peerperproduct = [1,5,8,6,8,4,5,1,7,3,8,3,5,2,8,3,2,6,1]
bins = [1,2,3,4,5,6,7,8,9]
plt.hist(peerperproduct,bins,histtype='bar',rwidth=0.5,label = 'population')
plt.xlabel('this is x axis for histogram')
plt.ylabel('this is y axis for histogram')
plt.title('Radom label for histogram')
plt.legend()
```

```python
myDays = [1,2,3,4,5]
```

```python
organic = [8,9,8,3,3]
ads = [8,9,7,12,8]
affiliate = [3,5,5,4,3]
seller = [9,5,8,9,14]
```

```python
plt.plot([],[],color = 'b',label = 'organic')
plt.plot([],[],color = 'g',label = 'ads')
plt.plot([],[],color = 'r',label = 'affiliate')
plt.plot([],[],color = 'c',label = 'seller')
plt.stackplot(myDays,organic,ads,affiliate,seller,colors = ['b','g','r','c'])
plt.xlabel('number of days')
plt.ylabel('sales')
plt.legend()
```
